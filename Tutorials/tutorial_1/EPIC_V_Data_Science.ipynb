```python
# NumPy: Operaciones avanzadas de forma (Reshape, Concatenación)

print("Ejemplos de manipulación de la forma de arrays")

# Creación de un array 1D
arr_1d = np.arange(12)
print("Array 1D original:", arr_1d)
# Reshape: Cambiar la forma de un array
# Convierte un array 1D de 12 elementos a un array 2D de 3 filas y 4 columnas
arr_2d = arr_1d.reshape((3, 4))
print("Array 2D (reshape a 3x4):\n", arr_2d)

# También podemos volver a un array 1D
arr_flat = arr_2d.flatten() # o arr_2d.ravel()
print("Array 1D (aplanado de nuevo):\n", arr_flat)
```

```python
print("\nEjemplos de Concatenación de Arrays")

# Creamos dos arrays para concatenar
arr_a = np.array([[1, 2], [3, 4]])
arr_b = np.array([[5, 6], [7, 8]])

print("Array A:\n", arr_a)
print("Array B:\n", arr_b)

# Concatenación vertical (por defecto axis=0)
# 'np.concatenate' es una función general; 'np.vstack' es un caso especial para concatenación vertical
concat_vertical = np.concatenate((arr_a, arr_b), axis=0) # Equivalente a np.vstack((arr_a, arr_b))
print("\nConcatenación vertical (np.concatenate con axis=0):\n", concat_vertical)


# Concatenación horizontal (axis=1)
# 'np.hstack' es un caso especial para concatenación horizontal
concat_horizontal = np.concatenate((arr_a, arr_b), axis=1) # Equivalente a np.hstack((arr_a, arr_b))
print("Concatenación horizontal (np.concatenate con axis=1):\n", concat_horizontal)

# Ejemplo de concatenar arrays de diferentes dimensiones (requiere ajuste si es necesario)
# Para concatenar arrays de diferentes dimensiones, a menudo se necesita usar np.newaxis
arr_c = np.array([10, 20, 30])
arr_d = arr_c[:, np.newaxis] # Transforma arr_c en una columna 2D
print("\nArray C (1D):", arr_c)
print("Array D (C como columna 2D):\n", arr_d)
print("Concatenación horizontal de C y D (ejemplo de diferente dimensión, requiere cuidado):\n", np.concatenate((arr_c[:, np.newaxis], arr_d), axis=1))
```

```python
# Pandas: Indexación explícita (.loc, .iloc) y Combinación de DataFrames

print("Ejemplos de Indexación con Pandas")

# Creamos un DataFrame de ejemplo
data_ciudades = {
    'Ciudad': ['Quito', 'Guayaquil', 'Cuenca', 'Ibarra', 'Loja'],
    'Pais': ['Ecuador', 'Ecuador', 'Ecuador', 'Ecuador', 'Ecuador'],
    'Poblacion': [3300000, 1600000, 800000, 700000, 350000],
    'Area_km2': [600, 100, 135, 140, 41]
}

df_ciudades = pd.DataFrame(data_ciudades, index=['UIO', 'GYE', 'CUE', 'IBA', 'LOJ'])
print("DataFrame de Ciudades:\n", df_ciudades)

```

```python
# .loc[]: Indexación basada en etiquetas (labels)
print("\nUtilizando .loc[] (por etiquetas)")
# Seleccionar una fila por su etiqueta de índice
print("Fila con índice 'GYE':\n", df_ciudades.loc['GYE'])

# Seleccionar una columna por su etiqueta
print("\nColumna 'Poblacion':\n", df_ciudades.loc[:, 'Poblacion']) # el ':' significa todas las filas

# Seleccionar un valor específico por fila y columna
print("\nPoblación de Ibarra (IBA):", df_ciudades.loc['IBA', 'Poblacion'])

# Seleccionar un rango de filas y columnas por etiqueta
print("\nFilas de UIO a IBA y columnas Ciudad y Poblacion:\n", df_ciudades.loc['UIO':'IBA', ['Ciudad', 'Poblacion']])
```

```python
# .iloc[]: Indexación basada en posiciones numéricas (integers)
print("\n--- Utilizando .iloc[] (por posiciones numéricas) ---")
# Seleccionar la primera fila (posición 0)
print("Primera fila (posición 0):\n", df_ciudades.iloc[0])

# Seleccionar una columna por su posición
print("\nSegunda columna (posición 1, 'Pais'):\n", df_ciudades.iloc[:, 1])

# Seleccionar un valor específico por posición de fila y columna
print("\nPoblación de la segunda ciudad (posición 1, columna 2):", df_ciudades.iloc[1, 2])

# Seleccionar un rango de filas y columnas por posición
print("\nPrimeras 2 filas y primeras 3 columnas:\n", df_ciudades.iloc[0:2, 0:3])
```

```python
print("\nEjemplos de Combinación de DataFrames con pd.concat")

# Creamos dos DataFrames para combinar
df_paises_1 = pd.DataFrame({
    'Pais': ['España', 'Francia'],
    'Capital': ['Madrid', 'París']
})

df_paises_2 = pd.DataFrame({
    'Pais': ['Alemania', 'Italia'],
    'Capital': ['Berlín', 'Roma']
})

print("DataFrame Paises 1:\n", df_paises_1)
print("DataFrame Paises 2:\n", df_paises_2)

# Concatenar verticalmente (por defecto, axis=0)
df_combinado_vertical = pd.concat([df_paises_1, df_paises_2], ignore_index=True)
print("\nDataFrame Combinado Verticalmente (con ignore_index=True):\n", df_combinado_vertical)

```

```python
print("\nEjemplo de Combinación de DataFrames con pd.merge")

# Reset index to create a column to merge on (for demonstration)
df_paises_1_reset = df_paises_1.reset_index().rename(columns={'index': 'Original_Index'})

# Create df_info_extra with a corresponding index column
df_info_extra_with_key = pd.DataFrame({
    'Moneda': ['EUR', 'EUR', 'EUR', 'EUR'],
    'Original_Index': [0, 1, 0, 1] # Assuming a key to merge on
})

print("DataFrame Paises 1 (con índice reseteado):\n", df_paises_1_reset)
print("DataFrame Info Extra (con clave para merge):\n", df_info_extra_with_key)

# Merge the two DataFrames on the 'Original_Index' column
df_merged_example = pd.merge(df_paises_1_reset, df_info_extra_with_key, on='Original_Index', how='left')

print("\nDataFrame Combinado con pd.merge:\n", df_merged_example)
```

| Característica                   | **NumPy**                                               | **Pandas**                                                                   |
| -------------------------------- | ------------------------------------------------------- | ---------------------------------------------------------------------------- |
| 🔢 **Tipo de estructura**        | Arrays multidimensionales (`ndarray`)                   | Series (1D) y DataFrames (2D, tipo tabla)                                    |
| 🧠 **Propósito principal**       | Cálculo numérico rápido, algebra lineal, matrices       | Manipulación y análisis de datos estructurados                               |
| ⚙️ **Velocidad de ejecución**    | Muy eficiente para operaciones numéricas intensivas     | Más versátil para manipular datos con etiquetas y columnas                   |
| 📊 **Manejo de datos faltantes** | Limitado, no diseñado para `NaN` o `None`               | Muy robusto: permite filtrar, rellenar, eliminar valores nulos               |
| 🧩 **Indexación**                | Por posición (índices numéricos)                        | Por etiquetas (columna, índice personalizado)                                |
| 🔄 **Funcionalidades comunes**   | Operaciones vectorizadas, funciones matemáticas         | Limpieza de datos, agrupaciones, fusiones, filtrado                          |
| 📈 **Visualización**             | No tiene soporte integrado                              | Se complementa bien con `matplotlib` y `seaborn`                             |
| 📦 **Ecosistema**                | Base para otras librerías como `pandas`, `scikit-learn` | Se integra muy bien con `NumPy`, `matplotlib`, `statsmodels`, `scikit-learn` |


**NumPy** es ideal cuando necesitas trabajar con grandes cantidades de datos numéricos en forma de vectores o matrices, con alto rendimiento computacional.

**Pandas** está diseñado para el análisis de datos estructurados (como archivos CSV, Excel o bases SQL), donde necesitas organizar, limpiar y explorar los datos de manera flexible.

## 5️⃣ Visualización avanzada con Seaborn y Matplotlib

```python
# Matplotlib: Estilo Orientado a Objetos (OO)
import matplotlib.pyplot as plt
print("Ejemplos de Matplotlib con Estilo Orientado a Objetos")

# Generar datos de ejemplo
# linspace es una función de la librería NumPy que crea una secuencia de números espaciados uniformemente dentro de un intervalo dado.
x = np.linspace(0, 10, 100)

# 1. Gráfico único con estilo OO
# plt.subplots() sin argumentos crea una figura y un solo eje (subplot)
fig, ax = plt.subplots(figsize=(8, 4)) # 'fig' es el objeto figura, 'ax' es el objeto ejes

# Plotear usando el método plot() del objeto 'ax'
ax.plot(x, np.sin(x), color='blue', linestyle='--', label='Seno(x)')

# Personalizar el gráfico usando los métodos del objeto 'ax'
ax.set_title('Función Seno (Estilo Orientado a Objetos)')
ax.set_xlabel('Valores de X')
ax.set_ylabel('Valores de Y')
ax.legend()
ax.grid(True)
plt.show()
```

```python
# 2. Múltiples subplots con estilo OO (como se menciona en el PDF)
# Creamos una figura y un array de objetos 'ax' (ej. 2 filas, 1 columna)
fig, ax = plt.subplots(2, 1, figsize=(8, 8), sharex=True) # sharex=True para compartir el eje X

# Accedemos a cada subplot a través del array 'ax'
ax[0].plot(x, np.sin(x), color='red')
ax[0].set_title('Gráfico Superior: Seno(x)')
ax[0].set_ylabel('Amplitud')
ax[0].grid(True)

ax[1].plot(x, np.cos(x), color='green')
ax[1].set_title('Gráfico Inferior: Coseno(x)')
ax[1].set_xlabel('Valores de X')
ax[1].set_ylabel('Amplitud')
ax[1].grid(True)

# Un título general para la figura
fig.suptitle('Comparación de Funciones Seno y Coseno', fontsize=16)

# Ajustar el diseño para evitar superposiciones
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajusta el subtítulo general
plt.show()
```

## ⚡ Data Scaling o Escalado de Datos

### 🎯 Objetivos:
- Mejorar el rendimiento de los algoritmos de aprendizaje automático

Muchos algoritmos, como regresión logística, SVM, KNN o redes neuronales, son sensibles a la escala de los datos. Si las variables tienen diferentes unidades o rangos, el modelo puede aprender de forma ineficiente o errónea.

- Acelerar la convergencia en algoritmos de optimización

Al escalar los datos, especialmente en modelos como redes neuronales o regresión lineal con gradiente descendente, se facilita que el proceso de entrenamiento sea más rápido y estable, evitando oscilaciones o pasos desbalanceados.

- Evitar que las características con mayor escala dominen el modelo

El escalado garantiza que ninguna variable tenga más peso que otras solo por su magnitud. Así se asegura que todas las variables contribuyan equitativamente al modelo y se mantenga una representación justa de los datos.

```python
# Importar las librerías necesarias
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el dataset Iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['species'] = df['target'].apply(lambda x: iris.target_names[x])
```

```python
# Mostrar las primeras filas del dataset
print("Dataset Iris original:")
display(df.head())
print("\nEstadísticas descriptivas del dataset original:")
display(df.describe())
```

### ¿Cuándo usar Normalización?

Transforma los datos para que todos estén entre un rango fijo, normalmente de 0 a 1.

✅ Ventajas:
Muy útil cuando necesitas que todos los datos estén entre 0 y 1.

Bueno para algoritmos basados en distancias (k-NN, redes neuronales).

⚠️ Desventajas:
Muy sensible a valores atípicos (outliers). Si hay un número muy grande o muy pequeño, puede distorsionar todo el escalado.


---


- Algoritmos que dependen de distancias (como k‑NN, k‑Means o redes neuronales), requieren que las características estén en una escala similar para evitar que las variables con magnitudes mayores dominen el cálculo de distancia.
- Si las características tienen rangos muy distintos (edad vs ingresos, por ejemplo), la normalización (Min‑Max) asegura que cada variable aporte proporcionalmente sin sesgos.
- También se recomienda cuando la distribución es desconocida o no es Gaussiana, pues no requiere media ni varianza conocidas.



---



```python
# Seleccionar una característica para visualizar antes y después de la transformación
feature_to_transform = 'sepal length (cm)'

#Aplicar Normalización (Min-Max Scaling)
print(f"\nAplicando Normalización (Min-Max Scaling) a '{feature_to_transform}'")
scaler_minmax = MinMaxScaler(feature_range=(0, 1))
df_normalized = df.copy()
df_normalized[feature_to_transform] = scaler_minmax.fit_transform(df[[feature_to_transform]])

print("\nEstadísticas descriptivas después de Normalización:")
display(df_normalized.describe())
```

### ¿Cuándo usar Estandarización?

Transforma los datos para que tengan media 0 y desviación estándar 1. Es decir, los convierte a una distribución normal estándar (también se llama Z-score scaling).


✅ Ventajas:
Bueno cuando los datos tienen una distribución normal.

Menos afectado por outliers que Min-Max.

⚠️ Desventajas:
Si los datos no tienen una forma de campana (no normal), el resultado puede no tener mucho sentido.


---



- Ideal para algoritmos que asumen distribución normal, como regresión logística, regresión lineal o PCA, donde se busca media cero y desviación estándar uno.
Funciona mejor si el conjunto de datos contiene outliers, ya que no se ve afectada por valores extremos como sí pasa con la normalización.

- También se utiliza en modelos con regularización (Lasso, Ridge, SVM, redes neuronales), donde el algoritmo penaliza coeficientes en función de su escala. Una estandarización adecuada mejora convergencia y estabilidad.

- Ayuda a acelerar el entrenamiento basado en descenso de gradiente al evitar que una característica domine las actualizaciones del gradiente.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python

# Separar las características (X) y el objetivo (y)
X = df.drop(['target', 'species'], axis=1)
y = df['target']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"\nTamaño del conjunto de entrenamiento: {X_train.shape}")
print(f"Tamaño del conjunto de prueba: {X_test.shape}")
```

```python
# Normalización (Min-Max Scaling)
scaler_minmax = MinMaxScaler(feature_range=(0, 1))
X_train_normalized = scaler_minmax.fit_transform(X_train)
X_test_normalized = scaler_minmax.transform(X_test) # Usamos transform en el test set

print("Datos de entrenamiento normalizados (primeras 5 filas):\n", X_train_normalized[:5])
print("\nDatos de prueba normalizados (primeras 5 filas):\n", X_test_normalized[:5])

# Estandarización (Standard Scaling)
scaler_standard = StandardScaler()
X_train_standardized = scaler_standard.fit_transform(X_train)
X_test_standardized = scaler_standard.transform(X_test) # Usamos transform en el test set

print("\nDatos de entrenamiento estandarizados (primeras 5 filas):\n", X_train_standardized[:5])
print("\nDatos de prueba estandarizados (primeras 5 filas):\n", X_test_standardized[:5])
```

##  🚨 Datos Faltantes

### 🎯 Objetivos:

En ciencia de datos, el manejo de datos faltantes es uno de los retos más comunes al trabajar con información real. Los datos incompletos pueden aparecer por errores de captura, problemas de almacenamiento o porque simplemente ciertos valores no existen. Ignorar estos valores puede llevar a análisis erróneos, modelos poco confiables o incluso fallos en los algoritmos. **¿Qué es un dato faltante? (`None`, `np.nan`, `pd.NA`)**
En Python existen varias formas de representar datos faltantes o valores nulos. Las más comunes son:

- `None`: Objeto nulo de Python, indica ausencia de valor.
- `np.nan`: Valor especial de tipo flotante proporcionado por NumPy, representa "Not a Number".
- `pd.NA`: Introducido en versiones recientes de pandas, es una forma general de representar datos faltantes para distintos tipos de datos.

**Concepto y aplicación de la matriz de sombras (shadow matrix)**

La **matriz de sombras** (o **shadow matrix**) es un concepto utilizado en el análisis de datos y estadísticas para representar la presencia o ausencia de datos en un conjunto de datos. Este tipo de matriz ayuda a visualizar y analizar los patrones de datos faltantes, facilitando el entendimiento de cómo y por qué los datos faltan.

**Concepto de Matriz de Sombras**

La matriz de sombras es una representación binaria del conjunto de datos original donde:

- **1** (o cualquier valor positivo) indica la presencia de un valor en esa posición del conjunto de datos original.

- **0** (o cualquier valor negativo) indica la ausencia de un valor en esa posición.

**Uso y Beneficios**

1. **Visualización de Datos Faltantes**:

- La matriz de sombras facilita la visualización de patrones en los datos faltantes. Puedes ver claramente qué filas o columnas tienen más datos faltantes y si hay patrones específicos.

2. **Análisis de Patrones de Datos Faltantes**:

- Ayuda a identificar si los datos faltantes están distribuidos aleatoriamente o si siguen algún patrón específico.

3. **Preprocesamiento**:

- La matriz de sombras se utiliza en técnicas de imputación para entender cómo los datos faltantes están relacionados con otras variables y para mejorar los métodos de imputación.

```python
(
    df_titanic
    #Genero una matriz de booleanos que me identifica con tru y false para que ,e puedan indicar cuales sno faltantes
    .isna()
    #Se remplaza para tener una matriz de datos con data más informativa
    .replace({
        False: "Not missing",
        True: "Missing"
    })
    # se debe tener nombres coordinados entre las variables originales y estas que se crean para ello se agrega el sufijo _NA
    .add_suffix("_NA")
    #concatenamos la matriz original con esta matriz de sommbras que acabamos de crear
    .pipe(
        lambda shadow_matrix: pd.concat(
            [df, shadow_matrix],
            axis="columns"
        )
    )
)
```

**Automatizando la generación de la matriz**

```python
df_titanic.missing.bind_shadow_matrix(only_missing = True) # only_missing = True toma en cuenta unicamente las variables que tienen valores faltantes
```
